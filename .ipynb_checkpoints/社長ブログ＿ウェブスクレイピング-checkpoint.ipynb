```python
my_session = requests.Session()
url = 'https://www3.ttc-net.co.jp/member-page/wp-login.php?'
r = my_session.post(url,data={'log':'r-ei', 'pwd':'Rong0903'})
bs = BeautifulSoup(r.content,'html.parser')
print(bs)
```

```python
tag_link = bs.find("li", id="menu-item-25613").find("a").get("href")
# print(tag_link)
url_boss = my_session.get(tag_link)
boss_bs = BeautifulSoup(url_boss.text,'html.parser')
print(boss_bs)
```

```python
page = boss_bs.find("div", id = "content")
for h2 in page.find_all("h2", class_= "entry-title"):
    boss_content_link = h2.find("a").get("href")
#     print(boss_content_link)
    boss_content_bs = BeautifulSoup(boss_content_link,'html.parser')
    print(boss_content_bs)

```
