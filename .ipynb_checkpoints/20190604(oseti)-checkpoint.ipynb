```python
import oseti
import sqlite3
from bs4 import BeautifulSoup
```

```python
con = None
#和数据库链接
con = sqlite3.connect("./restaurant.db")
# cursor是光标扫描的意思
cursor = con.cursor()
# 数据抽出
res = cursor.execute("SELECT review FROM REVIEWS WHERE restaurant_id ='1' ")
print(res)

# タプルから出します
content = cursor.fetchall()
type(content)
```
