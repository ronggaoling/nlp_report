```python

con = sqlite3.connect("./sample.db")
cursor = con.cursor()
res = cursor.execute("SELECT content FROM countries WHERE contry_name = 'アイスランド共和国'")
iceland = BeautifulSoup(cursor.fetchall()[0][0], "html.parser")
content = iceland.find("div", id="content").text
text = re.sub("[\n,\xa0,]","",content)
text = re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-…]+', "", text)
# print(text)
mecab_analysis(text)
# iceland_over=" ".join(mecab_analysis(text))
```

```python
con = sqlite3.connect("./sample.db")
cursor = con.cursor()
res = cursor.execute("SELECT content FROM countries WHERE contry_name = 'アイルランド'")
ireland = BeautifulSoup(cursor.fetchall()[0][0], "html.parser")
content = ireland.find("div", id="content").text
text = re.sub("[\n,\xa0,]","",content)
text = re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-…]+', "", text)
# print(text)
mecab_analysis(text)
ireland_over=" ".join(mecab_analysis(text))
```

```python
def mecab_analysis(texts):
    t = mc.Tagger("")
    t.parse('')
    output = []
    node =  t.parseToNode(texts)  
    while node:
        if node.surface != "":  # ヘッダとフッタを除外
            word_type = node.feature.split(",")[0]
            if word_type in ['名詞']:
                output.append(node.surface)
        node = node.next
        if node is None:
            break
    return output
```

```python
ice = iceland_over.split(" ")
ire = ireland_over.split(" ")
content = [ice,ire]
dictionary = Dictionary(content)
```

```python
# dictionary.filter_extremes(no_below=2, no_above=0.3)
corpus = [dictionary.doc2bow(doc) for doc in content]
lda = LdaModel(corpus, num_topics=2, id2word=dictionary,passes=10)
```

```python
for topic in lda.show_topics(num_topics=-1, num_words =10):
    print('topic id:{0[0]:d}, words={0[1]:s}'.format(topic))
```
