```python
driver_exe = '/mnt/c/Users/r-ei/Desktop/file/chromedriver.exe'
```

```python
# サイト内容のスクレイピングを始まり
song_url = "https://www.joysound.com/web/karaoke/ranking/foreign/monthly"
try:
    driver = webdriver.Chrome(driver_exe)
    driver.implicitly_wait(1)
    
    driver.get(song_url)
    song_bs = BeautifulSoup(driver.page_source, 'html.parser')
finally:
    driver.quit()
```

```python
# ランキングのテーブル
table = song_bs.find("div", class_="jp-cmp-ranking-list-003")
print(table)
```

```python
# 找到有歌词按钮地方的链接
space = table.find_all("div", class_='jp-cmp-table-column-001 jp-cmp-table-column-pt-001')
lyrics = []
for button in space:
    goal = button.find("li", class_= 'jp-cmp-music-lyrics-001 ng-scope')
    if goal is not None:
        goal=goal.find('a').get('href')
    else:
        goal = ""   
    lyrics.append(goal)  
```

```python
header = 'https://www.joysound.com'
lyrics_content = []

try:
    driver = webdriver.Chrome(driver_exe)
    driver.implicitly_wait(1)
    for need_links in lyrics:
        if need_links!= "":
            need_url =  header + need_links
#             requests.packages.urllib3.disable_warnings()
#             need_link = requests.get(header + need_url, verify=False)
#             need_link = requests.get(header + need_url)
            driver.get(need_url)
            time.sleep(1)
            need_linkbs = BeautifulSoup(driver.page_source, 'html.parser')
            lyric = need_linkbs.find('p', class_='ng-binding').text
        else:
            lyric = ""
        lyrics_content.append(lyric)
except Exception:
    pass
finally:
    driver.quit()
```

```python
conn = None

def connect():
    global conn
    conn = sqlite3.connect('./karaok.db')
    
def close():
    conn.close()
    
def create_table():
    conn.execute('DROP TABLE IF EXISTS singers')
    conn.execute('''CREATE TABLE singers(
    id               INTEGER PRIMARY KEY AUTOINCREMENT,
    singer            TEXT
    
    )''')
    conn.execute('DROP TABLE IF EXISTS songs')
    conn.execute('''CREATE TABLE songs(
    id               INTEGER PRIMARY KEY AUTOINCREMENT,
    singer_id         TEXT,
    name              TEXT,
    lytic             TEXT
    )''')
    

    
def load_singer(singer):
    conn.execute(
    'INSERT INTO singers (singer) VALUES (?)', (singer,))
    conn.commit()

def load_son(singer_id,name, lytic):
    conn.execute(
    'INSERT INTO songs (singer_id,name,lytic) VALUES (?,?,?)', (singer_id,name,lytic))
    conn.commit()
connect()    
create_table()
close()
```
