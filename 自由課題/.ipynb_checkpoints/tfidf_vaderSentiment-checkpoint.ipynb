```python
import re
import sqlite3
from gensim import corpora, matutils
from gensim.corpora.dictionary import Dictionary
from gensim.models.ldamodel import LdaModel
from sklearn.feature_extraction.text import TfidfVectorizer
from bs4 import BeautifulSoup
from wordcloud import WordCloud
from matplotlib import pyplot as plt
import pyLDAvis
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
import nltk
import math
import string
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize,word_tokenize
from collections import Counter
from nltk.stem.porter import *
from sklearn.feature_extraction.text import TfidfVectorizer
import MeCab as mc
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# 正規表現
wkachi_text = []
con = sqlite3.connect("./karaok.db")
cursor = con.cursor()
res = cursor.execute("SELECT lytic FROM songs")
for i in cursor.fetchall():
    cs = BeautifulSoup(i[0], 'html.parser')
#     print(cs.text)
   
    lytic_content = re.sub("[◆,●,...,-,(),◇]","",i[0])
    lytic_content = re.sub(r"\n"," ",lytic_content)
    lytic_content=lytic_content.lower()
    wkachi_text.append(lytic_content)
# print(wkachi_text[1])
```

```python
def mecab_analysis(texts):
    t = mc.Tagger("")
    t.parse('')
    output = []
    node =  t.parseToNode(texts)  
    while node:
        if node.surface != "":  # ヘッダとフッタを除外
            word_type = node.feature.split(",")[0]
            if word_type in ['名詞']:
                output.append(node.surface)
        node = node.next
        if node is None:
            break
    return output
```

```python
wagachi = []
for sentence in wkachi_text:
    magachi = mecab_analysis(sentence)
    wagachi.append(magachi)
# print(wagachi)
```

```python
result = []
del_word = ['the','t','to','i','ll','m','on','a',"-","'",'your','oh','and','at','you','re','be','my','come', 'baby','in','with','me',
           'isn','la','they','so','it','wo','o', 's','that','sha','when', 'was','d', 'em','all','we', 'ya','this','show','are','ve']
for word in wagachi:
    result_1 = []
    for w in word:       
        if w not in del_word:
            result_1.append(w)
        else:
            pass
    result.append(result_1)
print(result[1])  
```

```python
count_result = []
for content in result:
    count = Counter(content)
    count_result.append(count)
print(count_result[1])    
```

```python
# tf-idf https://blog.csdn.net/baimafujinji/article/details/51476117
def tf(word, count):
    return count[word] / sum(count.values())

def n_containing(word, count_list):
    return sum(1 for count in count_list if word in count)

def idf(word, count_list):
    return math.log(len(count_list) / (1 + n_containing(word, count_list)))

def tfidf(word, count, count_list):
    return tf(word, count) * idf(word, count_list)
```

```python
countlist = [count_result[1]]
for i, count in enumerate(countlist):
#     print("Top words in document {}".format(i + 1))
#     scores = {word: tfidf(word, count, countlist) for word in count}
    scores = {word: tfidf(word, count, countlist) for word in count}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
#     for word, score in sorted_words[:3]:
#         print("\tWord: {}, TF-IDF: {}".format(word, round(score, 5)))
print(sorted_words[:3])
print(sorted_words[:3][0][0])
```

```python
# Find special word in every song and put it into the atabase
outstanding_words = []
for lytic in count_result:
      
        for i, count in enumerate([lytic]):
            scores = {word: tfidf(word, count, [lytic]) for word in count}
            sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
            outstanding = sorted_words[:3][0][0]
            outstanding_words.append(outstanding)
print(outstanding_words)
conn = sqlite3.connect("./karaok.db")
c = conn.cursor()
c.execute(
"ALTER table songs ADD column special text")
conn.commit()
```

```python
for i,specialword in enumerate(outstanding_words):
    conn.execute(
    "update songs set special=? where id = ?", (specialword,i+1))
conn.commit()
conn.close()
```

```python
# 分析所有歌词的感情
emotion = []
analyzer = SentimentIntensityAnalyzer()
for lytic in wkachi_text:
    vs = analyzer.polarity_scores(lytic)
    emotion.append(vs)
# print(emotion[0]['compound'])
```

```python
# 把歌词的感情值都表示出来，并且放入database
emotion_results = []
for i in range(100):
    emotion_result = emotion[i]['compound']
    emotion_results.append(emotion_result)
print((emotion_results))
```

```python
conn = sqlite3.connect("./karaok.db")
c = conn.cursor()
c.execute(
"ALTER table songs ADD column emotion int")
conn.commit()
```

```python
for i,num in enumerate(emotion_results):
    conn.execute(
    "update songs set emotion=? where id = ?", (num,i+1))
conn.commit()
conn.close()
```
