```python
res = cursor.execute("SELECT review FROM REVIEWS WHERE restaurant_id ='1' ")
comment = cursor.fetchall()
wkachi_text = []
# 数据库里所有的东西都显示出来
for i in comment:
    text = re.sub("[\n,＼(^o^)／,(>_<),♪( ´θ｀)ノ,＼(^o^)／,⚪︎,（笑）]","",i[0])
#     text = re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-…]+', "", text)
    wkachi_text.append(text)
print(wkachi_text)
```

```python
def mecab_analysis(texts):
    t = mc.Tagger("")
    t.parse('')
    output = []
    node =  t.parseToNode(texts)  
    while node:
        if node.surface != "":  # ヘッダとフッタを除外
            word_type = node.feature.split(",")[0]
            if word_type in ['名詞']:
                output.append(node.surface)
        node = node.next
        if node is None:
            break
    return output
```

```python
wakachi_LD = []
for b in wkachi_text:
    wakachi_LD.append(mecab_analysis(b))
wakachi_LD
```

```python
stop_words = [ u'さん', u'をこと', u'さん', u'これ', u'今年', u'電話', u'英語', u'そう', u'そこ', u'こと', u'技術', u'社員',u'line',u'body',u'height']  
dictionary = Dictionary(wakachi_LD)
corpus = [dictionary.doc2bow(doc) for doc in wakachi_LD]
lda = LdaModel(corpus,num_topics=4,id2word=dictionary,passes=10)
for topic in lda.show_topics(num_topics=-1,num_words=10):
    print("topic id:{0[0]:d},words={0[1]}:s".format(topic))
```

```python
vis = pyLDAvis.gensim.prepare(lda,corpus,dictionary,sort_topics=False)
vis
```
